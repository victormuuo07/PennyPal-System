### **Feature Engineering**

1. Balance categories we are going to categorize the balance column into low, medium, high based on the distribution or percentiles
2. Age Groups - We can segment age into different groups such as youth, middle aged and senior

```python
balance_bins = [-float('inf'), df4['balance'].quantile(0.33), df4['balance'].quantile(0.66), float('inf')]
balance_labels = ['low', 'medium', 'high']
df4['balance_category'] = pd.cut(df4['balance'], bins=balance_bins, labels=balance_labels)
```

```python
age_bins = [0, 25, 50, 100]
age_labels = ['youth', 'middle-aged', 'senior']
df4['age-group'] = pd.cut(df4['age'], bins=age_bins, labels=age_labels)
```

```python
print(df4[['age', 'balance', 'balance_category', 'age-group']].head())
```

### **Modeling**

Will use K-means clustering algorithm to segment customers based on features such as age, balance, duration.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

features = df4[['age', 'balance', 'duration']]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

kmeans = KMeans(n_clusters=3, random_state=42)
df4['cluster'] = kmeans.fit_predict(scaled_features)

print(df4[['age', 'balance', 'duration', 'cluster']].head())

print("Cluster Centers", kmeans.cluster_centers_)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. Check KMeans metrics:
silhouette = silhouette_score(df4[['age','balance']], kmeans.labels_)
inertia = kmeans.inertia_

print(f'Silhouette score: {silhouette:.2f}')
print(f'Inertia: {inertia:.2f}')

```

```python
# 2. Visualize Cluster Centers:

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df4['age'], y=df4['balance'], hue=kmeans.labels_, palette='viridis',s=100, alpha=0.6 )
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='red', s=200, label='Cluster Centers')
plt.title('Cluster Centeres and Assignments')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.legend()
plt.show()
```

```python
# 3. Visualize Cluster Assignments (Age vs Balance)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df4['age'], y=df4['balance'], hue=kmeans.labels_, palette='viridis', s=100, alpha=0.6)
plt.title('Customer Segmentation (Cluster Assigments)')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.legend(title='Cluster')
plt.show()
```

```python
# 4. Elbow PLot 
inertia_values = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df4[['age', 'balance']])
    inertia_values.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia_values, marker='o', color='b', linestyle= '--')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```
